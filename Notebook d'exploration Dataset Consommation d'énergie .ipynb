```python
#Suppression de colonnes qui ne m'intéressent pas pour l'analyse
delcol = ['Opérateur',  
          'Nombre de mailles secretisées (agriculture)',
          'Nombre de mailles secretisées (industrie)', 
          'Nombre de mailles secretisées (secteur inconnu)',
          'Nombre de mailles secretisées (tertiaire)',
          'Nombre de mailles secretisées (résidentiel)', 
          'geom' ]

df = df.drop(columns=delcol)
```

```python
#Suppression d'années considérées comme trop lointaines pour l'analyse
annees_a_exclure = [2011, 2012, 2013, 2014, 2015]

# Filtrage des lignes correspondant aux années à supprimer
df_filtre = df[~df['Année'].isin(annees_a_exclure)]

# Sauvegarde du nouveau DataFrame dans un nouveau fichier CSV
df_filtre.to_csv('ecologie_powerbi.csv', index=False)
```

```python
# Quels sont les trois départements qui ont connu la plus grande baisse de consommation entre 2016 et 2021 ?
# Très bonne question 

#On commence par créer deux dataframes, l'un avec les données de 2016, l'autre avec celles de 2021
df_2016 = df_filtre[df_filtre['Année']==2016]
df_2021= df_filtre[df_filtre['Année']==2021]

#les dataframes ci-dessous regroupent les totaux de consommations par départements
total_consommation_par_departement_2016 = df_2016.groupby('Libellé Département')['Consommation totale (MWh)'].sum().reset_index()
total_consommation_par_departement_2021 = df_2021.groupby('Libellé Département')['Consommation totale (MWh)'].sum().reset_index()

#On renomme les consommations pour savoir à quelle année correspondront les colonnes de consommation après le merge
total_consommation_par_departement_2016 = total_consommation_par_departement_2016.rename(columns={'Consommation totale (MWh)': 'Consommation totale (MWh) 2016'})
total_consommation_par_departement_2021 = total_consommation_par_departement_2021.rename(columns={'Consommation totale (MWh)': 'Consommation totale (MWh) 2021'})

#On merge le tout 
total_consommation_par_departement_2016_2021 = pd.merge(total_consommation_par_departement_2016, total_consommation_par_departement_2021, on='Libellé Département', how='inner')

#Création d'une colonne affichant le taux d'évolution entre 2016 et 2021
total_consommation_par_departement_2016_2021['Evolution consommation'] = ((total_consommation_par_departement_2016_2021['Consommation totale (MWh) 2021']- total_consommation_par_departement_2016_2021['Consommation totale (MWh) 2016'])/total_consommation_par_departement_2016_2021['Consommation totale (MWh) 2016'])* 100

#C'est plus joli d'arrondir à deux chiffres après la virgule
total_consommation_par_departement_2016_2021['Evolution consommation'] = total_consommation_par_departement_2016_2021['Evolution consommation'].round(2)

#Tri des résultats de la colonne 'Evolution consommation' par ordre croissant
total_consommation_par_departement_2016_2021 = total_consommation_par_departement_2016_2021.sort_values(by='Evolution consommation', ascending=True)
total_consommation_par_departement_2016_2021 

# Nos grands gagnants sont donc le Loir-et-Cher, l'Eure-et-Loir et La Seine-Maritime, bravo à eux. 
#On ne félicite pas Mayotte, le Jura et les deux-Sèvres, en bas du classement 
```
